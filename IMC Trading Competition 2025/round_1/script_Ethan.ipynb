```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats 
```

```python
path = "data/"
price_day_0 = pd.read_csv(path + "prices_round_1_day_0.csv", delimiter=";")
price_day_minus_1 = pd.read_csv(path + "prices_round_1_day_-1.csv", delimiter=";")
price_day_minus_2 = pd.read_csv(path + "prices_round_1_day_-2.csv", delimiter=";")
trades_day_0 = pd.read_csv(path + "trades_round_1_day_0.csv", delimiter=";")
trades_day_minus_1 = pd.read_csv(path + "trades_round_1_day_-1.csv", delimiter=";")
trades_day_minus_2 = pd.read_csv(path + "trades_round_1_day_-2.csv", delimiter=";")
```

```python
prices = pd.concat([price_day_0, price_day_minus_1, price_day_minus_2]).reset_index(drop=True)
trades = pd.concat([trades_day_0, trades_day_minus_1, trades_day_minus_2]).reset_index(drop=True)
```

```python
prices
```

```python
sanity_check = len(prices) == len(price_day_minus_1)+len(price_day_minus_2)+len(price_day_0) 
if sanity_check:
    print("Sanity check passed")
else:
    print("Sanity check failed")
```

```python
trades
```

```python
sanity_check_bis = len(trades) == len(trades_day_minus_1)+len(trades_day_minus_2)+len(trades_day_0)
if sanity_check_bis:
    print("Sanity check passed")
else:
    print("Sanity check failed")
```
