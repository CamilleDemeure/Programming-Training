## **Backtest on one round**

```python
# To run the backtest, we need to specify the path to the script we want to run. 
# Use ../round_{i}/{name of the script} 
# At the end, add the number of the round we are in (so it uses the correct data from this round)
! prosperity3bt ../round_0/script.py 0
# It will display the profit for all assets that we can trade within this round. Here for example, we have the results for day -1 of round 0. If you use 1, it will display three results for day -2, day-1, and day 0. 
```

```python
# So, to backtest all round 1, you can use : 
! prosperity3bt ../round_2/kelp_bollinger_bands.py 2
```

## **Backtest on one day of a round**

```python
# You can also backtest only day 0 of round 1 : 
! prosperity3bt ../round_1/Trading_kelp_squid.py 1-0
```

## **Backtest on multiple days of a round**

```python
# If you want to backtest multiple days (day -1 and day 0) for one round : 
! prosperity3bt ../round_1/Trading_kelp_squid.py 1--1, 1-0
```

## **Backtest on multiple rounds**

```python
# For multiple rounds (imagine round 1 and round 2) : 
! prosperity3bt ../round_1/Trading_kelp_squid.py 1 2
```

## **Backtest on multiple days of one round by merging PnL across days**

```python
# To merge profit and loss across days of one round : 
! prosperity3bt example/starter.py 1 --merge-pnl
```

## **Backtest on one round with visualization**

```python
! prosperity3bt ../round_0/script.py 1 --vis 
# Open google chrome and show the results if you used the template_Trader for the script
```

Other possible commands

```python
# Automatically open the result in the visualizer when done
# Assumes your algorithm logs in the visualizer's expected format
$ prosperity3bt example/starter.py 1 --vis

# Write algorithm output to custom file
$ prosperity3bt example/starter.py 1 --out example.log

# Skip saving the output log to a file
$ prosperity3bt example/starter.py 1 --no-out

# Backtest on custom data
# Requires the value passed to `--data` to be a path to a directory that is similar in structure to https://github.com/jmerle/imc-prosperity-3-backtester/tree/master/prosperity3bt/resources
$ prosperity3bt example/starter.py 1 --data prosperity3bt/resources

# Print trader's output to stdout while running
# This may be helpful when debugging a broken trader
$ prosperity3bt example/starter.py 1 --print
```
